```python
# TWO METHOD OF EDA FOR NUMERIC DATA ATTRIBUTE
#define size of plot sheet 
plt.figure(figsize=(16,8)) 
#create swarmplot v
ax = sns.swarmplot(x="Age", y="PROT", data=df)
```

```python
plt.figure(figsize=(16, 8))
plt.title("COUNTPLOT FOR AGE")
#plot histogram based on age
sns.countplot(data = df, x = 'Age')
```

```python
# TWO METHOD OF EDA FOR NOMINAL DATA ATTRIUTE 
plt.figure(figsize=(16,8))
plt.title("CHOL FOR EACH CATEGORY")
sns.boxplot(x='Category',y='CHOL',data=df)
```

```python
df.groupby(["Category", "Sex"]).count()
```

```python
# BOX PLOT
plt.figure(figsize=(16,8))
sns.boxplot(x='Age', y='PROT', data=df)
```

```python
# BAR GRAPH
df['Sex'].value_counts().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Frequency");
```

```python
# HISTOGRAM
chm= df[df['Sex']=='m']['CHOL']

plt.figure(figsize=(16,8))
plt.title("MALE")
plt.hist(df[df['Sex']=='m']['CHOL'])
plt.xlabel('CHOL')
plt.ylabel('Number of PEOPLE')
plt.grid()
plt.show()
```

```python
# HISTOGRAM
chf= df[df['Sex']=='f']['CHOL']
plt.figure(figsize=(16,8))
plt.title("FEMALE")
plt.hist(df[df['Sex']=='f']['CHOL'])
plt.xlabel('CHOL')
plt.ylabel('Number of PEOPLE')
plt.grid()
plt.show()
```

```python
# SCATTER PLOT
plt.figure(figsize=(16,8))
plt.title("AGE vs CHOL")
plt.scatter(df['Age'],df['CHOL'])
plt.xlabel('Age')
plt.ylabel('CHOL')
plt.legend(loc='lower right')
```

```python
# SWAM PLOT
plt.figure(figsize=(16,8))
ax = sns.swarmplot(x="Sex", y="CHOL", data=df)
```

```python
# REG PLOT
plt.figure(figsize=(16,8))
plt.title("AGE vs PROT")
sns.regplot(x='Age',y='PROT',data=df)
```

```python
# VIOLIN PLOT
plt.figure(figsize=(16,8))
ax = sns.violinplot(x="Sex", y="PROT", data=df)
```
